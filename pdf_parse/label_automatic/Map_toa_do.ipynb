```python
path_input1 = "/content/drive/MyDrive/Resumes_Project/preprocess/dataset/annotation_vietnam/csv/"
path_input2 = "/content/drive/MyDrive/Resumes_Project/preprocess/dataset/annotation_vietnam/txt/"
path_output = "/content/drive/MyDrive/Resumes_Project/preprocess/dataset/annotation_vietnam/txt_sau_map/"
```

```python
def preprocess_file(file):
  print(file)
  df1 = pd.read_csv(path_input1 + file)

  # print(df1.head())
  

  data = np.array(df1)
  data = list(data)
  data1 = []
  for d in data:
    text = d[4]
    if text != '' and not pd.isnull(text):
      data1.append(list(d))

  # print(data1)
  file1 = file.replace("pdf", "pdf_")

  file2 = path_input2 + file1[:-3] + 'txt'
  if not os.path.exists(file2):
    print("not exist path")
    # continue
    return None
  f = open(file2, "r")
  data = f.read()

  # print(data)

  data= data.split("\n")

  data2 = []
  for d in data:
    d = d.split('\t')
    # print(d)
    if len(d) < 5:
      return None 
    text = d[4]
    label = d[5]
    data2.append([text, label])
  # print(data2)

  i = 0
  j = 0

  m = len(data1)
  n = len(data2)
  results = ""
  while ( i < m and j < n):
    d1 = data1[i]
    d2 = data2[j]
    text1 = d1[4]
    text2 = d2[0]
    if text1 == text2:
      i = i + 1
      j = j + 1
      text = text1
      xmin = d1[0]
      ymin = d1[1]
      xmax = d1[2]
      ymax = d1[3]
      label = d2[1]
      results += str(xmin) + "\t" + str(ymin) + "\t" + str(xmax) + "\t" + str(ymax) + "\t" + text + "\t" + label + "\n"
    else:
      i = i + 1

  fout = path_output + file.replace("pdf", "pdf-")[:-3] + "txt"
  f = open(fout, "w")
  f.write(results[: -1])
  f.close()
```

```python
files = os.listdir(path_input1)

for f in files:

  preprocess_file(f)
```
